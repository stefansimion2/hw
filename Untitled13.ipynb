```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('https://raw.githubusercontent.com/pointOfive/stat130chat130/refs/heads/main/CP/CSCS_data_anon.csv')

# Relevant columns for happiness and community participation
community_col = 'CONNECTION_activities_community_p3m'
happiness_col = 'WELLNESS_subjective_happiness_scale_score'

# Clean and prepare the data
data_clean = data[[community_col, happiness_col]].dropna()
data_clean[community_col] = pd.to_numeric(data_clean[community_col], errors='coerce')
data_clean[community_col] = data_clean[community_col].apply(lambda x: 'Yes' if x > 0 else 'No')

# Separate the happiness scores for 'Yes' and 'No' groups
yes_group = data_clean[data_clean[community_col] == 'Yes'][happiness_col].values
no_group = data_clean[data_clean[community_col] == 'No'][happiness_col].values

# Bootstrapping
n_iterations = 10000  # Number of bootstrap samples
bootstrap_diffs = []

# Resample and calculate mean difference
for _ in range(n_iterations):
    yes_sample = np.random.choice(yes_group, size=len(yes_group), replace=True)
    no_sample = np.random.choice(no_group, size=len(no_group), replace=True)
    mean_diff = np.mean(yes_sample) - np.mean(no_sample)
    bootstrap_diffs.append(mean_diff)

# Calculate the 95% confidence interval
lower_bound = np.percentile(bootstrap_diffs, 2.5)
upper_bound = np.percentile(bootstrap_diffs, 97.5)
mean_diff_observed = np.mean(yes_group) - np.mean(no_group)

# Display results
print("Observed Mean Difference:", mean_diff_observed)
print("95% Confidence Interval:", (lower_bound, upper_bound))

# Plot the bootstrap distribution
plt.figure(figsize=(10, 6))
plt.hist(bootstrap_diffs, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(lower_bound, color='red', linestyle='--', label='Lower CI Bound')
plt.axvline(upper_bound, color='red', linestyle='--', label='Upper CI Bound')
plt.axvline(mean_diff_observed, color='green', linestyle='-', label='Observed Mean Diff')
plt.title('Bootstrap Distribution of Mean Differences')
plt.xlabel('Mean Difference in Happiness')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
